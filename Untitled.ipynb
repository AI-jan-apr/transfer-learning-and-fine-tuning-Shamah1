## Data Preprocessing

```python
# encode string labels to integers
label_to_idx = {name: idx for idx, name in enumerate(class_names)}
idx_to_label = {idx: name for name, idx in label_to_idx.items()}

train_labels = [label_to_idx[p.split("/")[-2]] for p in train_paths]
val_labels   = [label_to_idx[p.split("/")[-2]] for p in val_paths]
test_labels  = [label_to_idx[p.split("/")[-2]] for p in test_paths]

print(f"Label mapping: {label_to_idx}")
```

### Augmentation Pipeline

```python
# augmentation for training only
data_aug = keras.Sequential([
    keras.layers.RandomFlip("horizontal"),
    keras.layers.RandomRotation(0.1),
    keras.layers.RandomZoom(0.1),
], name="augmentation")
```

### Dataloader

```python
# load, preprocess, and batch images
def load_image(path, label):
    image = tf.io.read_file(path)
    image = tf.image.decode_jpeg(image, channels=3)
    image = tf.image.resize(image, (IMG_SIZE, IMG_SIZE))
    image = preprocess_input(image)  # EfficientNet preprocessing
    label = tf.cast(label, tf.int32)
    return image, label

def make_dataset(paths, labels, training=False):
    ds = tf.data.Dataset.from_tensor_slices((paths, labels))
    if training:
        ds = ds.shuffle(len(paths), seed=SEED)
    ds = ds.map(load_image, num_parallel_calls=tf.data.AUTOTUNE)
    if training:
        ds = ds.map(lambda x, y: (data_aug(x, training=True), y),
                    num_parallel_calls=tf.data.AUTOTUNE)
    return ds.batch(BATCH_SIZE).prefetch(tf.data.AUTOTUNE)

train_ds = make_dataset(train_paths, train_labels, training=True)
val_ds   = make_dataset(val_paths, val_labels)
test_ds  = make_dataset(test_paths, test_labels)

print(f"Train batches: {len(train_ds)}")
print(f"Val   batches: {len(val_ds)}")
print(f"Test  batches: {len(test_ds)}")
```

### Visualize Sample Batch

## Experiment 1 — Feature Extraction

```python
# freeze base model, train classification head only
with strategy.scope():
    base_model = keras.applications.EfficientNetB0(
        include_top=False,
        weights="imagenet",
        input_shape=(IMG_SIZE, IMG_SIZE, 3)
    )
    base_model.trainable = False

    inputs  = keras.Input(shape=(IMG_SIZE, IMG_SIZE, 3))
    x       = base_model(inputs, training=False)
    x       = layers.GlobalAveragePooling2D()(x)
    x       = layers.Dropout(0.3)(x)
    outputs = layers.Dense(NUM_CLASSES, activation="softmax")(x)

    model_exp1 = keras.Model(inputs, outputs, name="EfficientNetB0_FeatureExtraction")
    model_exp1.compile(
        optimizer=keras.optimizers.Adam(learning_rate=1e-3),
        loss="sparse_categorical_crossentropy",
        metrics=["accuracy"]
    )

model_exp1.summary()
```

```python
callbacks_exp1 = [
    keras.callbacks.EarlyStopping(patience=3, restore_best_weights=True),
    keras.callbacks.ReduceLROnPlateau(patience=2, factor=0.2, min_lr=1e-7)
]

with mlflow.start_run(run_name="exp1-feature-extraction"):
    mlflow.log_params({
        "model"           : "EfficientNetB0",
        "approach"        : "feature_extraction",
        "epochs"          : EPOCHS,
        "batch_size"      : BATCH_SIZE,
        "lr"              : 1e-3,
        "trainable_params": 14091
    })

    # Train
    history_exp1 = model_exp1.fit(
        train_ds,
        validation_data=val_ds,
        epochs=EPOCHS,
        callbacks=callbacks_exp1
    )

    # Evaluate
    val_loss, val_acc = model_exp1.evaluate(val_ds, verbose=0)
    mlflow.log_metrics({"val_accuracy": val_acc, "val_loss": val_loss})

    # Plot
    plot_history(history_exp1, "Experiment_1_Feature_Extraction")
    mlflow.log_artifact("Experiment_1_Feature_Extraction.png")

    # Confusion Matrix
    y_pred = np.argmax(model_exp1.predict(test_ds), axis=1)
    y_true = np.concatenate([y.numpy() for _, y in test_ds])
    plt.figure(figsize=(12, 10))
    cm = confusion_matrix(y_true, y_pred)
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
                xticklabels=class_names, yticklabels=class_names)
    plt.title("Confusion Matrix - Experiment 1")
    plt.ylabel("Actual")
    plt.xlabel("Predicted")
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.savefig("exp1_confusion_matrix.png")
    plt.show()
    mlflow.log_artifact("exp1_confusion_matrix.png")

    # Classification Report
    print(classification_report(y_true, y_pred, target_names=class_names))

    # Log model
    mlflow.keras.log_model(model_exp1, "model_exp1")

    print(f"Val Accuracy: {val_acc:.4f}")
    print(f"Val Loss    : {val_loss:.4f}")
```

```python
model_exp1.save("exp1_feature_extraction.keras")
print(" Exp1 model saved")
```

## Experiment 2 — Fine-Tuning

```python
# unfreeze last 20 layers and recompile with lower learning rate
base_model.trainable = True

for layer in base_model.layers[:-20]:
    layer.trainable = False

trainable = sum([l.trainable for l in base_model.layers])
print(f"Trainable layers in base: {trainable}")

with strategy.scope():
    model_exp1.compile(
        optimizer=keras.optimizers.Adam(learning_rate=1e-5),
        loss="sparse_categorical_crossentropy",
        metrics=["accuracy"]
    )

print(" Model ready for fine-tuning")
```

```python
# train, plot, evaluate, and log all in one MLflow run
class MLflowCallback(keras.callbacks.Callback):
    def on_epoch_end(self, epoch, logs=None):
        mlflow.log_metrics({
            "train_accuracy": logs["accuracy"],
            "train_loss"    : logs["loss"],
            "val_accuracy"  : logs["val_accuracy"],
            "val_loss"      : logs["val_loss"]
        }, step=epoch)

callbacks_exp2 = [
    keras.callbacks.EarlyStopping(patience=3, restore_best_weights=True),
    keras.callbacks.ReduceLROnPlateau(patience=2, factor=0.2, min_lr=1e-7),
    MLflowCallback()
]

with mlflow.start_run(run_name="exp2-fine-tuning"):
    mlflow.log_params({
        "model"           : "EfficientNetB0",
        "approach"        : "fine_tuning",
        "epochs"          : EPOCHS,
        "batch_size"      : BATCH_SIZE,
        "lr"              : 1e-5,
        "unfrozen_layers" : 20
    })

    # Train
    history_exp2 = model_exp1.fit(
        train_ds,
        validation_data=val_ds,
        epochs=EPOCHS,
        callbacks=callbacks_exp2
    )

    # Evaluate
    val_loss, val_acc = model_exp1.evaluate(val_ds, verbose=0)
    mlflow.log_metrics({"val_accuracy": val_acc, "val_loss": val_loss})

    # Plot
    plot_history(history_exp2, "Experiment_2_Fine_Tuning")
    mlflow.log_artifact("Experiment_2_Fine_Tuning.png")

    # Confusion Matrix
    y_pred2 = np.argmax(model_exp1.predict(test_ds), axis=1)
    y_true2 = np.concatenate([y.numpy() for _, y in test_ds])

    plt.figure(figsize=(12, 10))
    cm2 = confusion_matrix(y_true2, y_pred2)
    sns.heatmap(cm2, annot=True, fmt='d', cmap='Blues',
                xticklabels=class_names, yticklabels=class_names)
    plt.title("Confusion Matrix - Experiment 2")
    plt.ylabel("Actual")
    plt.xlabel("Predicted")
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.savefig("exp2_confusion_matrix.png")
    plt.show()
    mlflow.log_artifact("exp2_confusion_matrix.png")

    # Log model
    mlflow.keras.log_model(model_exp1, "model_exp2")

    print(f"Val Accuracy: {val_acc:.4f}")
    print(f"Val Loss    : {val_loss:.4f}")
```

```python
# evaluate on test set
test_loss2, test_acc2 = model_exp1.evaluate(test_ds, verbose=0)
print(f"Test Accuracy: {test_acc2:.4f}")
print(f"Test Loss    : {test_loss2:.4f}")
```

## Analysis

### Feature Extraction vs Fine-Tuning

Fine-tuning outperformed feature extraction by ~1.6% on val accuracy (88.08% → 89.68%).
The improvement confirms that unfreezing the last 20 layers allowed the model
to better adapt EfficientNetB0 features to Food11.

### Generalization
Both models generalized well — test accuracy exceeded val accuracy in both
experiments, confirming clean data splits with no leakage.

### Convergence
Feature extraction converged steadily across 20 epochs with only 14,091 trainable parameters.
Fine-tuning required more epochs to stabilize — accuracy started at 74% in epoch 1
due to the newly unfrozen layers adapting, then improved consistently to 89.68%.

### Overfitting
No significant overfitting in either experiment.
In feature extraction, train and val stayed within 1% of each other.
In fine-tuning, the gap remained small throughout all 20 epochs,
supported by Dropout(0.3) and ReduceLROnPlateau callbacks.

### Weakest Class
Dairy product consistently showed the lowest recall across both experiments,
likely due to having the fewest training samples (429 train, 148 test).
